# Numpy Array Method
Numpy library is able to create a 2D space, in this case the values are still randomized. 
The array can be returned as follows

ArrayName[x_value][y_value]

You are able move and locate the index of the array by specifying the x-value and the y-value

```python
import numpy as np
```

```python
initValue = np.random.randint(100, size=(10, 10))
initValue
```

```python
# Getting mean function

def recursive(initValue):
    a = []
    for row in range(len(initValue)):
        # Iterates the column per row
        for value in range(len(initValue)):
            # Bottom left Corner
            if row == len(initValue)-1 and value == 0:
                _res = (
                    initValue[row][value] 
                    + initValue[row][value-1] # Left
                    + initValue[row-1][value] # Top
                    + initValue[row][value+1] # Right
                    + initValue[0][value] # Down
                )/5
            
            # Bottom right corner
            elif row == len(initValue)-1 and value == len(initValue)-1:
                _res = (
                    initValue[row][value]
                    + initValue[row][value-1] # Left
                    + initValue[row-1][value] # Top
                    + initValue[row][0] # Right
                    + initValue[0][value] # Down
                )/5

            # Top right corner
            elif row == 0 and value == len(initValue)-1:
                _res = (
                    initValue[row][value]
                    + initValue[row][value-1] # Left
                    + initValue[row-1][value] # Top
                    + initValue[row][0] # Right
                    + initValue[row+1][value] # Down
                )/5

            # Last column
            elif value == len(initValue)-1:
                _res = (
                    initValue[row][value]
                    + initValue[row][value-1] # Left
                    + initValue[row-1][value] # Top
                    + initValue[row][0] # Right
                    + initValue[row+1][value] # Down
                )/5

            # Last Row
            elif row == len(initValue)-1:
                _res = (
                    initValue[row][value]
                    + initValue[row][value-1] # Left
                    + initValue[row-1][value] # Top
                    + initValue[row][value+1] # Right
                    + initValue[0][value] # Down
                )/5
            else:
                _res = (
                    initValue[row][value]
                    + initValue[row][value-1] # Lef
                    + initValue[row-1][value] # Top
                    + initValue[row][value+1] # Right
                    + initValue[row+1][value]
                )/5
            a.append(_res)
    return a
```

```python
newMatt = np.around(np.array(recursive(initValue)).reshape(len(initValue), len(initValue)), decimals=0)
newMatt
```

```python
import matplotlib.pyplot as plt


initValue = newMatt
newMatt = np.around(np.array(recursive(initValue)).reshape(len(initValue), len(initValue)), decimals=0)
newMatt

fig, ax = plt.subplots()
im = ax.imshow(newMatt)

# Loop over data dimensions and create text annotations.
for i in range(len(initValue)):
    for j in range(len(initValue)):
        text = ax.text(j, i, newMatt[i, j],
                    ha="center", va="center", color="w")

ax.set_title("Dendritic Cell Vis")
fig.tight_layout()
# plt.show()
```
